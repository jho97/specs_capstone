```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv('../marijuana_tax/washington_marijuana_cleaned.csv')
df
```

```python
df = pd.DataFrame(data=df)
df
```

```python
#dropna with a how parameter to drop rows with all elements missing
df.dropna(how='all', inplace=True)
df
```

```python
df.info()
```

```python
#drop the tax percentage
df.drop(['Weighted average local sales tax rate6'], axis=1, inplace=True)
df
```

```python
# regex replace for the commas
df.replace({'\,':''}, regex = True, inplace= True)
df
```

```python
# fill in the - with a 0
df.replace({'\-': 0}, regex = True, inplace= True)
df
```

```python
df.info()
```

```python
df.rename(columns = {"Quarter of Activity2":"quarter", "Taxable retail sales as reported by cannabis collectives (before tax)3":"marijuana_before_tax", "Estimated retail value of cannabis and cannabis products subject to state and local sales taxes":"marijuana_after_tax", "Estimated state sales tax collections":"state_tax_collection"}, inplace=True)
df
```

```python
# I dropping columns that I might not use
df.drop([
         "Estimated retail value of cannabis and cannabis products sold at a licensed cannabis retailer (before tax)4",
         "Total retail value of cannabis and cannabis products (before tax)",
         "Sales exempt from sales tax for medical purposes5", 
         "Estimated local sales tax collections"], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
df['marijuana_before_tax'] = df['marijuana_before_tax'].astype(str).astype(int)
df['marijuana_after_tax'] = df['marijuana_after_tax'].astype(str).astype(int)
df['state_tax_collection'] = df['state_tax_collection'].astype(str).astype(int)
df.info()
```

```python
df = df.drop("quarter", axis=1)
df
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
names = df.columns
d = scaler.fit_transform(df)
scaled_df = pd.DataFrame(d, columns=names)
scaled_df
```
