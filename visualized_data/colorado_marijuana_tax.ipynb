```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.read_csv("../marijuana_tax/colorado_marijuana_cleaned.csv")
df
```

```python
#creat a dataframe
df = pd.DataFrame(data=df)
```

```python
#drop NAN values
df.dropna(inplace=True)
df
```

```python
# identifying the rows that I want to drop
# noticing that the number columns need to be converted to int
df.info()
```

```python
# cleaning the $ and , in order to convert the Dtypes
df.replace({'\$':'', '\,':'', '\(':'', '\)':''}, regex = True, inplace= True)
df.head
```

```python
df.head()
```

```python
#converting objects to int
#RMS_tax_total has an unexpected space after the name
df['state_sales_ta_total'] = df['state_sales_ta_total'].astype(str).astype(int)
df['medical_marijuana'] = df['medical_marijuana'].astype(str).astype(int)
df['retail_marijuana'] = df['retail_marijuana'].astype(str).astype(int)
df['RMS_tax_total '] = df['RMS_tax_total '].astype(str).astype(int)
df['RMS_local_government_distribution'] = df['RMS_local_government_distribution'].astype(str).astype(int)
df['RMS_retained_by_state_total'] = df['RMS_retained_by_state_total'].astype(str).astype(int)
df['RMS_marijuana_tax_cash_fund'] = df['RMS_marijuana_tax_cash_fund'].astype(str).astype(int)
df['RMS_public_school_fund'] = df['RMS_public_school_fund'].astype(str).astype(int)
df['RMS_general_fund'] = df['RMS_general_fund'].astype(str).astype(int)
df['RME_tax_total'] = df['RME_tax_total'].astype(str).astype(int)
df['RME_public_school_capital_construction_assistance_fund'] = df['RME_public_school_capital_construction_assistance_fund'].astype(str).astype(int)
df['RME_public_school_fund'] = df['RME_public_school_fund'].astype(str).astype(int)
df['license_and_fees_total'] = df['license_and_fees_total'].astype(str).astype(int)
df['total_taxes_and_fees'] = df['total_taxes_and_fees'].astype(str).astype(int)
```

```python
#all the Dtypes have changed to what I want them to be
df.info()
```

```python
# renaming 'RMS_tax_total ' to take away the space
df.rename(columns = {"RMS_tax_total ":"RMS_tax_total", "state_sales_ta_total" : "state_sales_tax_total"}, inplace=True)
df
```

```python
# dropping total years from the revenue month
df_taxes_months = df.drop(index=[12, 25, 38, 51, 64, 77, 90, 103])
df_taxes_months 
```

```python
df_taxes_months = df_taxes_months.drop("Revenue Month", axis=1)
df_taxes_months.reset_index()
```
