# Merge Both Datasets

#### https://pandas.pydata.org/docs/user_guide/merging.html (figure 11)

```python
merged_df = pd.concat([washington_df, idaho_df], axis=1).reindex(washington_df.index)
merged_df
```

```python
# https://www.geeksforgeeks.org/get-a-list-of-a-particular-column-values-of-a-pandas-dataframe/
# Making the values in the columns to a list
washington_values = merged_df["washington_dollars_millions"].tolist()
# washington_values
```

```python
idaho_values = merged_df["idaho_dollars_millions"].tolist()
# idaho_values
```

```python
# Getting the index values to a list
time_values = merged_df.index.tolist()
# time_values
```

```python
# https://www.geeksforgeeks.org/plot-a-vertical-line-in-matplotlib/
vert_line = time_values[38]
# vert_line
```

### Little explanation
### I used the vert_line variable to attach 2014 Q3, x-point and so that the vertical line in the graph shows the point of after the legalization of marijuana and before legalization
#### https://www.tutorialspoint.com/how-do-you-plot-a-vertical-line-on-a-time-series-plot-in-pandas#:~:text=Using%20panda%20we%20can%20create,%2C%20linestyle%3D%22dashed%22. Used this to look at adding that vertical line

```python
# https://www.tutorialspoint.com/how-to-make-two-plots-side-by-side-using-python 

from matplotlib import pyplot as plt
xPoints = time_values
y1Points = washington_values
y2Points = idaho_values

plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.plot(xPoints, y1Points)
plt.axvline(x= vert_line)
plt.title("Washington")
plt.xlabel('X-axis ')
plt.ylabel('Y-axis ')

plt.subplot(1, 2, 2) # index 2
plt.plot(xPoints, y2Points)
plt.axvline(x= vert_line)
plt.title("Idaho")
plt.xlabel('X-axis ')
plt.ylabel('Y-axis ')

plt.show()
```
